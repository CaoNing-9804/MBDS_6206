```python
def find_loyal(dataframe,col):
    after_sort = dataframe.sort_values(by = col, ascending=False)
    cut_off = after_sort[col].sum()*0.5
    add = 0
    order = 0
    for i in range(len(after_sort)):
        if add < cut_off:
            add +=  after_sort[col].iloc[i]
            order += 1
        else:
            break
    user = after_sort['user_id'].iloc[:order]
    same = after_sort[after_sort[col]==after_sort[col].iloc[order]]
    user2 = same['user_id']
    user_id = list(set(user.values) | set(user2.values))
    
    
    return user_id
```

```python
ot = find_loyal(df,'order_times')
os = find_loyal(df,'order_spend')
st = find_loyal(df,'sub_times')
c = find_loyal(df,'cancelled')
l = find_loyal(df,'login')
vp = find_loyal(df,'view_product')
at = find_loyal(df,'add_item_to_cart')
pa = find_loyal(df,'points_added')
pd = find_loyal(df,'points_deducted')
f = find_loyal(df,'formdata')
r = find_loyal(df,'review')
```

```python
def lists_union(*lists):
    return list(set.union(*map(set, lists)))
```

```python
fund_list = lists_union(ot,os,st)
```

```python
len(fund_list)
```

```python
infor_list = lists_union(l,vp,at,pa,pd,f,r)
```

```python
len(infor_list)
```

```python
def inner_elements(list1, list2):
    return list(set(list1) & set(list2))
```

```python
fund_infor_list = inner_elements(fund_list, infor_list)
```

```python
len(fund_infor_list)
```

```python
def move_elements(list1, list2):
    for element in list(list1):
        if element in list2:
            list1.remove(element)
    return list1
```

```python
user_fund = move_elements(fund_list, fund_infor_list)
```

```python
len(user_fund)
```

```python
user_infor = move_elements(infor_list, fund_infor_list)
```
