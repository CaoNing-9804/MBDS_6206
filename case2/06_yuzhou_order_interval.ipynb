```python
def preprocess_orders(file_path,time_interval):
    df = pd.read_csv(file_path)
    df['date_placed'] = pd.to_datetime(df['date_placed']).dt.date
    df['min_date'] = df.groupby('user_id')['date_placed'].transform('min')
    df['date_difference'] = (df['date_placed'] - df['min_date']).dt.days
    df = df[(df['date_difference'] <= time_interval) | (df['date_placed'] == df['min_date'])]
    df.drop(['min_date', 'date_difference'], axis=1, inplace=True)
    df.reset_index(drop=True, inplace=True)
    df_grouped = df.groupby('user_id').size().reset_index(name=f'order_times_in_{time_interval}days')
    return df_grouped
```

```python
def preprocess_cancel_orders(file_path,time_interval):
    df = pd.read_csv(file_path)
    df['date_placed'] = pd.to_datetime(df['date_placed']).dt.date
    df['min_date'] = df.groupby('user_id')['date_placed'].transform('min')
    df['date_difference'] = (df['date_placed'] - df['min_date']).dt.days
    df = df[(df['date_difference'] <= time_interval) | (df['date_placed'] == df['min_date'])]
    df.drop(['min_date', 'date_difference'], axis=1, inplace=True)
    df.reset_index(drop=True, inplace=True)
    df_grouped = df.groupby('user_id').size().reset_index(name=f'cancel_order_times_in_{time_interval}days')
    return df_grouped
```
