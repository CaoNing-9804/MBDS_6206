```python
import pandas as pd
import numpy as np
```

```python
review = pd.read_csv('D:/6206/Hook Coffee data Sep 2022/Review-2022-09-15.csv')
```

```python
form = pd.read_csv('D:/6206/Hook Coffee data Sep 2022/FormData-2022-09-15.csv')
```

```python
buy_comment = pd.DataFrame(columns=['user_id','buy_comment'])
```

```python
buy_comment['user_id']=review['user_id'].value_counts().index
```

```python
buy_comment['buy_comment']=review['user_id'].value_counts().values
```

```python
buy_comment
```

```python
cancel_comment = pd.DataFrame(columns=['user_id','cancel_comment'])
```

```python
cancel_comment['user_id']=form['user_id'].value_counts().index
cancel_comment['cancel_comment']=form['user_id'].value_counts().values
```

```python
cancel_comment
```

```python
comment = buy_comment.merge(cancel_comment, how='outer')
```

```python
comment = comment.fillna(value=0)
```

```python
comment['total_comment']=comment['buy_comment']+comment['cancel_comment']
```

```python
comment=comment.rename(columns={'buy_comment':"review_comment",'cancel_comment':"form_comment"})
```

```python
comment
```
