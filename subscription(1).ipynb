```python
list1 = ['product_id', 'shipping_postal_code', 'country', 'interval',
       'brewing_method','shipping_method', 'currency']

list2 = ['user_id','sub_times','product_id_max_item','product_id_max_freq', 'shipping_postal_code_max_item','shipping_postal_code_max_freq','country_max_item','country_max_freq', 'interval_max_item','interval_max_freq',
       'brewing_method_max_item','brewing_method_max_freq','shipping_method_max_item','shipping_method_max_freq', 'currency_max_item','currency_max_freq','active','cancelled','error','paused','first_create','last_update','period']

res = pd.DataFrame(columns=list2)

for i in sub['user_id'].value_counts().index:
    df = sub[(sub['user_id'] == i)]
    new = []
    
    times = len(df)
    new.append(i)
    new.append(times)
    for j in list1:
        stat = df[j].value_counts()
        try:
            freq = stat.values[0]
        except:
            freq = np.nan
        try:
            item = stat.idxmax()
        except:
            item = np.nan
        
        new.append(item)
        new.append(freq)
        
    status = df['status'].value_counts()
    try:
        active = status['active']
    except:
        active = 0
        
    try:
        cancelled = status['cancelled']
    except:
        cancelled = 0
        
    try:
        error = status['error']
    except:
        error = 0
        
    try:
        paused = status['paused']
    except:
        paused = 0
        
    first_create = df['created'].iloc[-1]
    last_update = df['updated'].iloc[0]
    period = last_update - first_create
    
    new.append(active)
    new.append(cancelled)
    new.append(error)
    new.append(paused)
    new.append(first_create)
    new.append(last_update)
    new.append(period)
    
    new = pd.Series(new,index=list2)
    df1 = pd.DataFrame(columns=list2)
    df1 = df1.append(new,ignore_index=True)
    
    res = pd.concat([res,df1])
```
